```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
from IPython.display import HTML, display
import pandas as pd
import math
```

```python
weather = pd.read_csv("Aus weather.csv",encoding='latin-1')
weather['year']=pd.DatetimeIndex(weather['Date/Time']).year
weather2=weather.loc[weather['year']==2019]

weather2.head()
```

```python
# Create a map
m_2 = folium.Map(location=[-33.75339911776345, 151.08014306634533], tiles='cartodbpositron', zoom_start=13)

# Add points to the map
for idx, row in weather2.iterrows():
    Marker([row['Latitude'], row['Longitude']]).add_to(m_2)

# Display the map
m_2
```

```python
# Create the map
m_3 = folium.Map(location=[-33.79221167884241, 151.02521142571882], tiles='cartodbpositron', zoom_start=13)

# Add points to the map
mc = MarkerCluster()
for idx, row in weather.iterrows():
    if not math.isnan(row['Longitude']) and not math.isnan(row['Latitude']):
        mc.add_child(Marker([row['Latitude'], row['Longitude']]))
m_3.add_child(mc)

# Display the map
m_3
```

```python
# Create a base map
m_4 = folium.Map(location=[-33.79221167884241, 151.02521142571882], tiles='cartodbpositron', zoom_start=5)

def color_producer(val):
    if val == 'Rain':
        return 'forestgreen'
    elif val == 'Hail':
        return 'blue'
    elif val == ' Waterspout':
        return 'black'
    elif val == 'Tornado':
        return 'orange'
    elif val == 'Lighting':
        return 'darkred'
    else:
        return 'darkyellow'

# Add a bubble map to the base map, weather2 for 2019 data
for i in range(0,len(weather2)):
    Circle(
        location=[weather2.iloc[i]['Latitude'], weather2.iloc[i]['Longitude']],
        radius=20,
        color=color_producer(weather2.iloc[i]['Database'])).add_to(m_4)

# Display the map
m_4
```

```python
weather1=weather.loc[weather['Database'].isin(['Rain'])]
weather1.tail()
```

```python
# Create a base map
m_5 = folium.Map(location=[-33.79221167884241, 151.02521142571882], tiles='cartodbpositron', zoom_start=12)

# Add a heatmap to the base map - Rain
HeatMap(data=weather1[['Latitude', 'Longitude']], radius=10).add_to(m_5)

# Display the map
m_5
```
